```python
import pandas as pd
import numpy as np

df = pd.read_csv("data/hotel.csv")

print (df.shape)

df.head(10)
```

```python
print(df["Travel type"].unique())
print(df["Hotel Class"].unique())
```

```python
class_2_hotels = df.loc[df.loc[:, "Hotel Class"] == "Class 2"]
class_3_hotels = df.loc[df.loc[:, "Hotel Class"] == "Class 3"]
class_4_hotels = df.loc[df.loc[:, "Hotel Class"] == "Class 4"]

solo_hotels = df.loc[df.loc[:, "Travel type"] == "solo"]
couple_hotels = df.loc[df.loc[:, "Travel type"] == "couple"]
friends_hotels = df.loc[df.loc[:, "Travel type"] == "friends"]
family_hotels = df.loc[df.loc[:, "Travel type"] == "family"]
business_hotels = df.loc[df.loc[:, "Travel type"] == "business"]

hotels = [class_2_hotels, class_3_hotels, class_4_hotels, solo_hotels, couple_hotels, friends_hotels, family_hotels, business_hotels]
names = ["class_2", "class_3", "class_4", "solo", "couple", "friends", "family", "business"]
```

```python
# 독립변수 service features에 관한 감정 점수
def make_xdata (data):    
    x_data=[] 
    for revec in data:
        x_data.append(revec[2:-1])        
        
    return np.array(x_data)

# 종속변수 star ratings에서 만족을 1 불만족을 0로 설정

def make_y_label (data):
    y_2la=[]
    
    for revec in data:        
        if revec[-1]>=40:
            y_2la.append(1)
        else:
            y_2la.append(0)
            
    return np.array(y_2la)

inputs = []
all_ys = []
for i in hotels:
  inputs.append(make_xdata(i.values.tolist()))
  all_ys.append(make_y_label(i.values.tolist()))

'''X variables'''
all_input=make_xdata(df.values.tolist())

'''Y variable'''
all_y=make_y_label(df.values.tolist())

print(all_input[0])
print(all_y[0])
```

```python
# K fold cross validation: K=5
from sklearn.model_selection import KFold

def K_fold_data (data, k):
    
    kf=KFold(n_splits=k, random_state=7, shuffle=True)
    K_train_data=[]
    K_validation_data=[]
    
    for train_index, validation_index in kf.split(data):
        fold_train_set, fold_validation_set = data[train_index], data[validation_index]
        K_train_data.append(fold_train_set)
        K_validation_data.append(fold_validation_set)
        
    return np.array(K_train_data), np.array(K_validation_data)

'''X 데이터의 train/test split'''
all_x_train, all_x_test = K_fold_data(all_input,5)

'''Y 데이터의 train/test split'''
all_y_train,all_y_test = K_fold_data(all_y,5)

all_x_trains, all_x_tests, all_y_trains, all_y_tests = [], [], [], []
for i in range(len(inputs)):
  all_x = K_fold_data(inputs[i],5)
  all_x_trains.append(all_x[0])
  all_x_tests.append(all_x[1])
  all_y_ = K_fold_data(all_ys[i],5)
  all_y_trains.append(all_y_[0])
  all_y_tests.append(all_y_[1])

print(len(all_x_train),len(all_x_train[0]),len(all_x_test[0]))
print(len(all_y_train[0]),len(all_y_test[0]))
```

## **2.2 예측 모델 건설**

독립변수: 서비스 특징에 관한 만족도,
종속변수: 호텔 서비스 전체 만족도

```python
# 선형 모델인 Logit 예측 모델

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score
import pickle

def logitRegression(x_train,x_test,y_train,y_test):
    f1_train=[]
    f1_test=[]
    models=[]
    
    for i in range(len(x_train)):
        logit=LogisticRegression()
        logit=logit.fit(x_train[i],y_train[i])

        models.append(logit)
                
        y_pre_train=logit.predict(x_train[i])
        y_pre_test=logit.predict(x_test[i])  

        f1_train.append(f1_score(y_train[i],y_pre_train))
        f1_test.append(f1_score(y_test[i],y_pre_test))
        
    return models,f1_train,f1_test

log_models,log_f1_train,log_f1_test = logitRegression(all_x_train,all_x_test,all_y_train,all_y_test)

log_models_hotels, log_f1_trains, log_f1_tests = [], [], []
for i in range(len(all_x_trains)):
  x, y, z = logitRegression(all_x_trains[i], all_x_tests[i], all_y_trains[i], all_y_tests[i])
  log_models_hotels.append(x)
  log_f1_trains.append(y)
  log_f1_tests.append(z)

print(log_models)
print(log_f1_train)
print(log_f1_test)
```

```python
# 예측 모델 성능 결과
from scipy.stats import sem

# print("F1: ", np.mean(log_f1_train),sem(log_f1_train),np.mean(log_f1_test),sem(log_f1_test))
for i in range(len(log_f1_trains)):
  print("F1: ", np.mean(log_f1_trains[i]), sem(log_f1_trains[i]), np.mean(log_f1_tests[i]), sem(log_f1_tests[i]))
```

```python
#Coefficients 확인, Coefficients의 절대값은 feature importance라고 볼 수 있음
# print(log_models[0].coef_)
for i in range(len(log_models_hotels)):
  print(log_models_hotels[i][0].coef_)
```

```python
# 비선형 모델인 lightGBM 예측 모델

from lightgbm import LGBMClassifier
import joblib
from sklearn.model_selection import GridSearchCV

def LGBM(x_train,x_test,y_train,y_test):
    f1_train=[]
    f1_test=[]
    models=[]
    
    for i in range(len(x_train)):
        lgbm=LGBMClassifier(objective='binary')
        lgbm=lgbm.fit(x_train[i],y_train[i])

        models.append(lgbm)
        
        y_pre_train=lgbm.predict(x_train[i])
        y_pre_test=lgbm.predict(x_test[i])        

        f1_train.append(f1_score(y_train[i],y_pre_train))
        f1_test.append(f1_score(y_test[i],y_pre_test))
        
    return models,f1_train,f1_test

LGBM_models, LGBM_f1_train, LGBM_f1_test = LGBM(all_x_train,all_x_test,all_y_train,all_y_test)
LGBM_models_hotels, LGBM_f1_trains, LGBM_f1_tests = [], [], []
for i in range(len(all_x_trains)):
  x, y, z = LGBM(all_x_trains[i], all_x_tests[i], all_y_trains[i], all_y_tests[i])
  LGBM_models_hotels.append(x)
  LGBM_f1_trains.append(y)
  LGBM_f1_tests.append(z)

print(LGBM_models)
print(LGBM_f1_train)
print(LGBM_f1_test)
```

```python
# 예측 모델 성능 결과
from scipy.stats import sem

for i in range(len(LGBM_f1_trains)):
  print("F1: ", np.mean(LGBM_f1_trains[i]), sem(LGBM_f1_trains[i]), np.mean(LGBM_f1_tests[i]), sem(LGBM_f1_tests[i]))
# print("F1: ", np.mean(LGBM_f1_train),sem(LGBM_f1_train),np.mean(LGBM_f1_test),sem(LGBM_f1_test))
```
